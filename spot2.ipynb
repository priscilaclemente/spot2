```python
# Función para determinar el valor de 'type_new'
def determinar_type_new(type_transaction):
    type_transaction = set(type_transaction)  # Convertir a un conjunto para obtener valores únicos
    if type_transaction == {'renta'}:
        return 'renta'
    elif type_transaction == {'venta'}:
        return 'venta'
    elif type_transaction == {'renta', 'venta'}:
        return 'renta/venta'
    return 'desconocido'  # Para identificar casos diferentes
```

```python
# Agrupar por 'id' y aplicar la función para determinar 'type_new'
df_prices_new = df_prices.groupby('id').agg(
    type_new=('type_transaction', determinar_type_new),
    price_area=('price_area', 'first'),
    currency_type=('currency_type', 'first'),
    rate=('rate', 'first'),
    max_rate=('max_rate', 'first'),
    spot_id=('spot_id', 'first'),
    created_at_prices=('created_at_prices', 'first'),
    updated_at_prices=('updated_at_prices', 'first'),
    deleted_at_prices=('deleted_at_prices', 'first')
).reset_index()
```

```python
# Definir la función de safe_divide
def safe_divide(numerator, denominator):
    return numerator / denominator if denominator != 0 else np.nan
```

```python
# Crear la nueva columna rate_mx_m2
df_final['rate_mx_m2'] = df_final.apply(
    lambda row: (
        safe_divide(row['rate'], row['square_space']) * 18.46 if row['price_area'] == 1 and row['currency_type'] == 2 else
        safe_divide(row['rate'], row['square_space']) if row['price_area'] == 1 and row['currency_type'] == 1 else
        row['rate'] * 18.46 if row['price_area'] == 2 and row['currency_type'] == 2 else
        row['rate'] if row['price_area'] == 2 and row['currency_type'] == 1 else
        np.nan
    ),
    axis=1
)
```

```python
# Crear la nueva columna rate_mx_m2
df_final['rate_mx_m2_max'] = df_final.apply(
    lambda row: (
        safe_divide(row['max_rate'], row['square_space']) * 18.46 if row['price_area'] == 1 and row['currency_type'] == 2 else
        safe_divide(row['max_rate'], row['square_space']) if row['price_area'] == 1 and row['currency_type'] == 1 else
        row['max_rate'] * 18.46 if row['price_area'] == 2 and row['currency_type'] == 2 else
        row['max_rate'] if row['price_area'] == 2 and row['currency_type'] == 1 else
        np.nan
    ),
    axis=1
)
```

```python
def convert_longitude(longitude):
    if isinstance(longitude, str):
        # Eliminar el símbolo de grados y la dirección
        longitude = longitude.replace('°', '').strip()
        
        # Separar el valor numérico de la dirección
        parts = longitude.split()
        value = float(parts[0])
        direction = parts[1]
    else:
        # Si ya es un número, simplemente retornarlo
        value = float(longitude)
        direction = ''  # Asumimos que la dirección no está presente si ya es un número
    
    # Convertir a formato decimal
    if direction.upper() in ['W', 'WEST']:
        value = -value
    
    return value

# Aplicar la función a toda la columna 'longitude'
df_final['longitude_decimal'] = df_final['longitude'].apply(convert_longitude)
```

```python
def convert_latitude(latitude):
    if isinstance(latitude, str):
        # Eliminar el símbolo de grados y la dirección
        latitude = latitude.replace('°', '').strip()
        
        # Separar el valor numérico de la dirección
        parts = latitude.split()
        value = float(parts[0])
        direction = parts[1]
    else:
        # Si ya es un número, simplemente retornarlo
        value = float(latitude)
        direction = ''  # Asumimos que la dirección no está presente si ya es un número
    
    # Convertir a formato decimal
    if direction.upper() in ['S', 'SOUTH']:
        value = -value
    
    return value

# Aplicar la función a toda la columna 'longitude'
df_final['latitude_decimal'] = df_final['latitude'].apply(convert_latitude)
```
